```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
import matplotlib.pyplot as plt

#Load Model and Image
model = tf.keras.models.load_model('/Users/anudavis/Desktop/Assignment/my_model.weights.best.hdf5') 
    #'path_to_your_hdf5 model
img = image.load_img('/Users/anudavis/Desktop/Assignment/PredictingImages/1547090.jpg', target_size=(150, 150))  
    #'path_to_your_image.jpg, 'The target_size should match the input size for your model

#Prediction
plt.imshow(img)
plt.show()
x=image.img_to_array(img)
x=np.expand_dims(x,axis=0)
images=np.vstack([x])
pred=model.predict(images,batch_size=1) 
if pred[0][0]>0.5:
    print("ApplePie")
elif pred[0][1]>0.5:
    print("BagelSandwich")
elif pred[0][2]>0.5:
    print("Bibimbop")
elif pred[0][3]>0.5:
    print("Bread")
elif pred[0][4]>0.5:
    print("FriedRice")
elif pred[0][5]>0.5:
    print("Pork")
else:
    print("Unknown")
```
